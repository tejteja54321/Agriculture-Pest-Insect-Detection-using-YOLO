## Dataset visualization

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset folders
train_dir = r'E:/TEJA/proj7/code/dataset/train/images'
valid_dir = r'E:/TEJA/proj7/code/dataset/valid/images'
test_dir = r'E:/TEJA/proj7/code/dataset/test/images'

# Count number of image files
num_train = len([f for f in os.listdir(train_dir) if os.path.isfile(os.path.join(train_dir, f))])
num_valid = len([f for f in os.listdir(valid_dir) if os.path.isfile(os.path.join(valid_dir, f))])
num_test  = len([f for f in os.listdir(test_dir) if os.path.isfile(os.path.join(test_dir, f))])

# Data for plotting
splits = ['Train', 'Validation', 'Test']
counts = [num_train, num_valid, num_test]

# Use seaborn style
sns.set(style="whitegrid")

# Plot
plt.figure(figsize=(7, 5))
sns.barplot(x=splits, y=counts, palette='viridis')

plt.title('Number of Images vs Data Split')
plt.ylabel('Number of Images')
plt.xlabel('Data Split')

# Add numbers on top of bars
for i, count in enumerate(counts):
    plt.text(i, count + max(counts)*0.01, str(count), ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.show()
```

```python
import os
from PIL import Image
from collections import Counter

# Dataset folders
image_dirs = [
    r'E:/TEJA/proj7/code/dataset/train/images',
    r'E:/TEJA/proj7/code/dataset/valid/images',
    r'E:/TEJA/proj7/code/dataset/test/images'
]

aspect_ratios = []

# Collect aspect ratios from all images
for folder in image_dirs:
    for filename in os.listdir(folder):
        file_path = os.path.join(folder, filename)
        if os.path.isfile(file_path):
            try:
                img = Image.open(file_path)
                w, h = img.size
                ratio = round(w / h, 2)  # Round to 2 decimal places for grouping
                aspect_ratios.append(ratio)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

# Count frequencies
ratio_counts = Counter(aspect_ratios)

# Prepare data for plotting
ratios = list(ratio_counts.keys())
counts = list(ratio_counts.values())

# Sort by aspect ratio
ratios, counts = zip(*sorted(zip(ratios, counts)))

# Plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
sns.barplot(x=list(ratios), y=list(counts), palette='viridis')

plt.title('Image Frequency vs Aspect Ratio')
plt.xlabel('Aspect Ratio (Width / Height)')
plt.ylabel('Number of Images')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python

# Dataset folders
image_dirs = [
    r'E:/TEJA/proj7/code/dataset/train/images',
    r'E:/TEJA/proj7/code/dataset/valid/images',
    r'E:/TEJA/proj7/code/dataset/test/images'
]


widths = []
heights = []

# Collect widths and heights from all images
for folder in image_dirs:
    for filename in os.listdir(folder):
        file_path = os.path.join(folder, filename)
        if os.path.isfile(file_path):
            try:
                img = Image.open(file_path)
                w, h = img.size
                widths.append(w)
                heights.append(h)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

# Count frequencies
width_counts = Counter(widths)
height_counts = Counter(heights)

# Sort by width/height
sorted_widths = sorted(width_counts.keys())
width_freqs = [width_counts[w] for w in sorted_widths]

sorted_heights = sorted(height_counts.keys())
height_freqs = [height_counts[h] for h in sorted_heights]

# Plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

plt.plot(sorted_widths, width_freqs, marker='o', color='blue', label='Width')
plt.plot(sorted_heights, height_freqs, marker='o', color='green', label='Height')

plt.title('Image Frequency vs Width and Height')
plt.xlabel('Pixels')
plt.ylabel('Number of Images')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np

widths = []
heights = []

# Collect widths and heights
for folder in image_dirs:
    for filename in os.listdir(folder):
        file_path = os.path.join(folder, filename)
        if os.path.isfile(file_path):
            try:
                img = Image.open(file_path)
                w, h = img.size
                widths.append(w)
                heights.append(h)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

# Count frequencies
width_counts = Counter(widths)
height_counts = Counter(heights)

# Use same x-axis (unique sorted pixel sizes)
all_sizes = sorted(set(width_counts.keys()).union(height_counts.keys()))

width_freqs = [width_counts.get(size, 0) for size in all_sizes]
height_freqs = [height_counts.get(size, 0) for size in all_sizes]

# Set seaborn style
sns.set(style="whitegrid")

# Bar width and positions
x = np.arange(len(all_sizes))
bar_width = 0.4

plt.figure(figsize=(12, 6))

# Bars
plt.bar(x - bar_width/2, width_freqs, width=bar_width, label='Width', color='blue')
plt.bar(x + bar_width/2, height_freqs, width=bar_width, label='Height', color='green')

plt.xlabel('Pixel Value (Width / Height)')
plt.ylabel('Number of Images')
plt.title('Image Frequency vs Width and Height')
plt.xticks(x, all_sizes, rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
